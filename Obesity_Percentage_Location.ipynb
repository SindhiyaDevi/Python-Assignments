```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
df = pd.read_csv("Obesity")
df
```

```python
df = df.drop(['YearEnd','LocationDesc', 'Datasource','Class', 'Topic','Question', 'Data_Value_Unit', 'Data_Value_Type', 'Data_Value_Alt', 'Data_Value_Footnote_Symbol', 'Data_Value_Footnote','Total', 'GeoLocation', 'ClassID', 'TopicID', 'QuestionID', 'DataValueTypeID', 'StratificationCategory1', 'Stratification1', 'StratificationCategoryId1', 'StratificationID1'], axis=1)
df
```

```python
df.isnull().sum()
```

```python
df =df.drop(["Low_Confidence_Limit"], axis=1)
df
```

```python
df = df.drop(df.columns[[3]], axis=1)
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python


Age_map = {'18 - 24':0, '25 - 34':1, '35 - 44':2,'45 - 54':3, '55 - 64':4, 
          '65 or older':5}

df['Age(years)'] = df['Age(years)'].map(Age_map)

df
```

```python
edu_map = {'less than high school': 0, 'high school graduate':1, 'some college or technical school':2,'college graduate':3}
edu_map
df['Education'] = df['Education'].map(edu_map)
df
```

```python
income_map = {'less than $15,000':0, '$15,000 - $24,999':1, '$25,000 - $34,999':2,'$35,000 - $49,999':3, '$50,000 - $74,999':4, 
          '$75,000 or greater':5, 'data not reported':6}

df['Income'] = df['Income'].map(income_map)
df
```

```python
data = df.drop(columns =['Gender', 'Race/Ethnicity', 'LocationAbbr'])
data.dropna(subset=['Data_Value', 'YearStart'], inplace=True)
data.isna()
```

```python
correlation = data[["YearStart", "Data_Value", "LocationID"]]
corr =correlation.corr()
corr
```

```python
X = data[['YearStart', 'Data_Value']]
y =data[['LocationID']]
X.dropna(how='all')
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```
